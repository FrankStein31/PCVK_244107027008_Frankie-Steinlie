```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg","https://imgs.search.brave.com/EWkFkPyA2jEez7iv9846E5uT2CyXfsldMyRIouQgdTU/rs:fit:860:0:0:0/g:ce/aHR0cHM6Ly9wYW5n/b3ZldC5jb20vd3At/Y29udGVudC91cGxv/YWRzLzIwMjQvMDYv/Q3V0ZS1raXR0ZW4t/ZHJvb3BpbmctZWFy/LXdpdGgtZ3JheS1m/dXItd2Fsa2luZ19C/cm9rZW4tV2hpdGVf/U2h1dHRlcnN0b2Nr/LmpwZw"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url) #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #convert image to RGB
  final_frame = cv.hconcat((image, image_2)) #concatenate image
  cv2_imshow(final_frame) #display image
  print('\n')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlopen

def baca_url(url):
    resp = urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    return image

url = "https://imgs.search.brave.com/EWkFkPyA2jEez7iv9846E5uT2CyXfsldMyRIouQgdTU/rs:fit:860:0:0:0/g:ce/aHR0cHM6Ly9wYW5n/b3ZldC5jb20vd3At/Y29udGVudC91cGxv/YWRzLzIwMjQvMDYv/Q3V0ZS1raXR0ZW4t/ZHJvb3BpbmctZWFy/LXdpdGgtZ3JheS1m/dXItd2Fsa2luZ19C/cm9rZW4tV2hpdGVf/U2h1dHRlcnN0b2Nr/LmpwZw"

original_image = baca_url(url)

def display_image(img, title):
    plt.figure(figsize=(10, 6))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

display_image(original_image, 'Original Foto')
```

```python
# Buat garis vertikal dan garis menyilang diagonal pada image keluaran
def garis(img):
    height, width = img.shape[:2]
    img_copy = img.copy()
    # Vertical
    cv2.line(img_copy, (width//2, 0), (width//2, height), (255, 255, 255), 2)
    # Diagonal
    cv2.line(img_copy, (0, 0), (width, height), (255, 255, 255), 2)
    return img_copy

image_tugas2 = garis(original_image)
display_image(image_tugas2, 'Tugas 2: Garis Vertikal and Diagonal')
```

```python
# Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang tertentu
def garis_hori(img, length=200):
    height, width = img.shape[:2]
    img_copy = img.copy()
    start_point = (width//2 - length//2, height//2)
    end_point = (width//2 + length//2, height//2)
    cv2.line(img_copy, start_point, end_point, (255, 255, 255), 2)
    return img_copy

image_tugas3 = garis_hori(original_image)
display_image(image_tugas3, 'Tugas 3: Garis Horizontal')
```

```python
# Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar
def random_kotak(img, box_size=50):
    height, width = img.shape[:2]
    img_copy = img.copy()
    x = np.random.randint(0, width - box_size)
    y = np.random.randint(0, height - box_size)
    img_copy[y:y+box_size, x:x+box_size] = (255, 255, 255)
    return img_copy

image_tugas4 = random_kotak(original_image)
display_image(image_tugas4, 'Tugas 4: Random Box Putih')
```

**TUGAS (GARIS KTP)**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlopen

def baca_ktp(url):
    resp = urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    return image

url = "https://blogpictures.99.co/bisniscom-7.jpg"

original_image = baca_ktp(url)
modified_image = original_image.copy()

def display_image(img, title):
    plt.figure(figsize=(12, 8))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

display_image(original_image, 'KTP Asli')

def cover_area(img, top_left, bottom_right, color1, color2):
    x1, y1 = top_left
    x2, y2 = bottom_right

    mid_x = (x1 + x2) // 2

    cv2.rectangle(img, (x1, y1), (mid_x, y2), color1, -1)
    cv2.rectangle(img, (mid_x, y1), (x2, y2), color2, -1)

# Menutup area NIK (sesuaikan koordinat berdasarkan gambar KTP)
cover_area(modified_image, (220, 185), (620, 225), (255, 0, 0), (0, 255, 0))

# Menutup area Nama (sesuaikan koordinat berdasarkan gambar KTP)
cover_area(modified_image, (220, 225), (620, 265), (0, 0, 255), (255, 255, 0))

display_image(modified_image, 'KTP Dimodifikasi')

print("KTP asli dan KTP yang dimodifikasi telah ditampilkan secara terpisah.")
print("OpenCV menggunakan format warna BGR, sedangkan Matplotlib menggunakan RGB.")
print("Oleh karena itu, kita mengkonversi BGR ke RGB saat menampilkan dengan Matplotlib.")
print("Saat menggunakan cv2.imshow(), konversi ini tidak diperlukan.")
```
